```python
def get_genres(dataframe):
    return (dataframe['genre_urls'].replace("/genres/","").replace("|"," | ").replace("-"," "))
```

```python
get_genres(df)[0]
```

```python
df['gen'] = df.apply(get_genres,axis=1)
```

```python
df.rename(columns={'gen':'Genres'},inplace=True)
```

```python
#seems that some books were written in negative years!
#print and observe

df[df.year < 0]

#all negative are BC!

def BC_yn(df):
    if df['year'] < 0:
        return "Yes"
    else:
        return "No"
```

```python
df['BC (YES/NO)'] = df.apply(BC_yn,axis=1)
```
